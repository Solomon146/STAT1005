### Question 5 (2 pt)

Example:
* input: [1,2,3,4,5]
* output: [4, 3, 2]

### Question 7 (3 pt)

1. Given a dictionary in Python, where all values are numeric, write a function called `get_values_average` to find the average of all values in the dictionary.

Example:
* input: {"a": 1, "b": 2, "c":3}
* output: 2.0

````python
def get_values_average(a_dict):
  # Calculate sum
  total = sum(a_dict.values())
  
  # Calculate average
  average = total/len(a_dict)

  return average

get_values_average({"a": 1, "b": 2, "c":3})
````

2. Given a dictionary, where all values are numeric, write a function called `get_argmin_key` to extract the key whose value is the smallest. You can assume that there is only one key with the smallest value.

Example:
* input: {"a": 1, "b": 2, "c":3}
* output: "a"

````python
def get_argmin_key(a_dict):
  # Get the min value
  min_value = min(a_dict.values())
  
  # Get the key which with the min value
  min_key = None
  min_key = [key for key, value in a_dict.items() if value==min_value]
 
  # Your code ends here
  return min_key

get_argmin_key({"a": 1, "b": 2, "c": 3})
````
